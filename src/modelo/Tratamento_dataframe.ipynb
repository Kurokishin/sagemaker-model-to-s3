# **Tratamento do Dataframe**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

### Lendo o dataframe

```python
df = pd.read_csv("dataframe/Hotel Reservations.csv")
```

```python
df
```

```python
def roomType(type):
    if type == "Room_Type 1":
        return 1
    elif type == "Room_Type 2":
        return 2
    elif type == "Room_Type 3":
        return 3
    elif type == "Room_Type 4":
        return 4
    elif type == "Room_Type 5":
        return 5
    elif type == "Room_Type 6":
        return 6
    else:
        return 7
```

```python
df['room_type_reserved'] = df['room_type_reserved'].apply(roomType)
```

```python
figura = plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True);
```

```python
df.drop(columns = ['Booking_ID', 'type_of_meal_plan', 'booking_status', 'market_segment_type', 'required_car_parking_space', 'no_of_weekend_nights', 'no_of_week_nights', 'repeated_guest', 'lead_time', 'no_of_previous_cancellations', 'no_of_previous_bookings_not_canceled', 'arrival_month', 'arrival_date'], axis = 1, inplace = True)
```

### Inserindo a coluna "label_avg_price_per_room" e relacionando-a com a coluna "avg_price_per_room"

```python
def avg_set_labels(avg):
    if avg <= 85:
        return 1
    elif 85 < avg < 115:
        return 2
    else:
        return 3
```

```python
df['label_avg_price_per_room'] = df['avg_price_per_room'].apply(avg_set_labels)
```

```python
df
```

### Exclusão da coluna "avg_price_per_room"

```python
df.pop("avg_price_per_room")
```

```python
df = pd.get_dummies(df, prefix = ['room_type_reserved', 'arrival_year', 'no_of_adults', 'no_of_children', 'no_of_special_requests'],
                   columns = ['room_type_reserved', 'arrival_year', 'no_of_adults', 'no_of_children', 'no_of_special_requests'])
```
