# **Tratamento do Dataframe**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv("../dataframe/Hotel Reservations.csv")
```

```python
df
```

```python
df.insert(19, "label_avg_price_per_room", 0)
```

```python
for i in range(len(df['avg_price_per_room'])):
    if df.at[i, "avg_price_per_room"] <= 85.0:
        df.at[i, "label_avg_price_per_room"] = 1
    elif 85.0 < df.at[i, "avg_price_per_room"] < 115:
        df.at[i, "label_avg_price_per_room"] = 2
    else:
        df.at[i, "label_avg_price_per_room"] = 3
```
